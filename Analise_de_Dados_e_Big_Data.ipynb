```python
#Histograma em Python com a biblioteca Matplotlib, que oferece uma variedade de funções para visualização de dados.


import matplotlib.pyplot as plt

# Dados de exemplo (números aleatórios)
dados = [1, 3, 5, 2, 3, 7, 8, 6, 10, 12, 14, 15, 17, 18, 20]

# Criar o histograma
plt.hist(dados, bins=5, color='blue', edgecolor='black')  # bins define o número de intervalos no histograma

# Adicionar rótulos e título
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.title('Histograma')

# Exibir o histograma
plt.show()
```

```python
#Exemplo Poligono de Frequência


import matplotlib.pyplot as plt

# Dados de exemplo (frequências de classes)
classes = [1, 2, 3, 4, 5]  # Classes
frequencias = [5, 10, 15, 20, 25]  # Frequências correspondentes

# Calcular os pontos para o polígono de frequência
x = classes
y = frequencias

# Criar o polígono de frequência
plt.plot(x, y, marker='o', linestyle='-')

# Adicionar rótulos e título
plt.xlabel('Classes')
plt.ylabel('Frequências')
plt.title('Polígono de Frequência')

# Exibir o polígono de frequência
plt.grid(True)
plt.show()
```

```python
#Exemplo Gráfico de Barra

import matplotlib.pyplot as plt

# Dados de exemplo (categorias e valores)
categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 15, 25]

# Cores para as barras
cores = ['blue', 'green', 'red', 'orange']

# Criar o gráfico de barras
plt.bar(categorias, valores, color=cores)

# Adicionar rótulos e título
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Gráfico de Barras com Cores')

# Exibir o gráfico
plt.show()
```

```python
#Exemplo de Gráfico em Coluna

import matplotlib.pyplot as plt

# Dados de exemplo
categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 15, 25]

# Criar o gráfico de colunas
plt.bar(categorias, valores, color='skyblue')

# Adicionar rótulos e título
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Gráfico de Colunas')

# Exibir o gráfico
plt.show()
```

```python
#Exemplo de Gráfico de Setor

import matplotlib.pyplot as plt

# Dados de exemplo
labels = ['A', 'B', 'C', 'D']
valores = [30, 20, 25, 25]

# Calcular a porcentagem total
total = sum(valores)

# Calcular porcentagens
porcentagens = [(valor / total) * 100 for valor in valores]

# Criar o gráfico de setor
plt.pie(porcentagens, labels=labels, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral', 'lightskyblue'])

# Adicionar título
plt.title('Gráfico de Setor')

# Exibir o gráfico
plt.show()
```

```python
#Exemplo Gráfico de Linha

import matplotlib.pyplot as plt

# Dados de exemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Criar o gráfico de linha
plt.plot(x, y, marker='o', linestyle='-')

# Adicionar rótulos e título
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Linha')

# Exibir o gráfico
plt.grid(True)
plt.show()
```

```python
#Exemplo de Pirâmide Etária

import matplotlib.pyplot as plt

# Dados de exemplo para a pirâmide etária (população por faixa etária)
idades = ['0-4', '5-9', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40-44', '45-49']
populacao_masculina = [100, 120, 130, 110, 90, 80, 70, 60, 50, 40]
populacao_feminina = [95, 115, 125, 105, 85, 75, 65, 55, 45, 35]

# Criar o gráfico de pirâmide etária
plt.figure(figsize=(10, 6))

# Plotar a população masculina
plt.barh(idades, populacao_masculina, color='skyblue', label='Masculino')
# Plotar a população feminina com valores negativos para inverter o eixo y
plt.barh(idades, [-x for x in populacao_feminina], color='pink', label='Feminino')

# Adicionar rótulos e título
plt.xlabel('População')
plt.ylabel('Faixa Etária')
plt.title('Pirâmide Etária')

# Adicionar legenda
plt.legend()

# Exibir o gráfico
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Adiciona uma grade apenas no eixo x
plt.show()
```

```python
#Exemplo piramide etária por sexo entre 2000 e 2010 na paraiba



idades = ['0-4', '5-9', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40-44', '45-49']
populacao_masculina_2000 = [100, 120, 130, 110, 90, 80, 70, 60, 50, 40]
populacao_feminina_2000 = [95, 115, 125, 105, 85, 75, 65, 55, 45, 35]

populacao_masculina_2010 = [110, 130, 140, 120, 100, 90, 80, 70, 60, 50]
populacao_feminina_2010 = [105, 125, 135, 115, 95, 85, 75, 65, 55, 45]
```

```python
import matplotlib.pyplot as plt

# Criar o gráfico de pirâmide etária
plt.figure(figsize=(10, 6))

# Plotar a população masculina de 2000
plt.barh(idades, populacao_masculina_2000, color='blue', label='Masculino - 2000')
# Plotar a população feminina de 2000 com valores negativos para inverter o eixo y
plt.barh(idades, [-x for x in populacao_feminina_2000], color='pink', label='Feminino - 2000')

# Plotar a população masculina de 2010
plt.barh(idades, populacao_masculina_2010, left=populacao_masculina_2000, color='lightblue', label='Masculino - 2010')
# Plotar a população feminina de 2010 com valores negativos para inverter o eixo y
plt.barh(idades, [-x for x in populacao_feminina_2010], left=[-x for x in populacao_feminina_2000], color='lightpink', label='Feminino - 2010')

# Adicionar rótulos e título
plt.xlabel('População')
plt.ylabel('Faixa Etária')
plt.title('Pirâmide Etária - Paraíba (2000 vs 2010)')

# Adicionar legenda
plt.legend()

# Exibir o gráfico
plt.grid(axis='x', linestyle='--', alpha=0.7)  # Adiciona uma grade apenas no eixo x
plt.show()
```

```python
#Exemplo Gráfico Polar

import matplotlib.pyplot as plt
import numpy as np

# Dados de exemplo
theta = np.linspace(0, 2*np.pi, 100)
r = np.abs(np.sin(2*theta))

# Criar o gráfico polar
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)  # Criar um subplot polar

# Plotar os dados
ax.plot(theta, r)

# Adicionar título
plt.title('Gráfico Polar')

# Exibir o gráfico
plt.show()
```
